```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn
%matplotlib inline
```

```python
x = np.linspace(0,10,100)
y1 = np.exp(-2*np.abs(x-5))
plt.figure(figsize=[10,5])
plt.subplot(1,2,1)
y1_plot, = plt.plot(x,y1)
y2 = np.ones(100)*0.1
y2_plot, = plt.plot(x,y2,color = 'red')
plt.legend([y1_plot,y2_plot],['Integrand','Uniform Density'])
plt.subplot(1,2,2)
y1_plot2, = plt.plot(x,y1)
y3 = (1/np.sqrt(2*np.pi))*np.exp(-0.5*(x-5)**2)
y3_plot, = plt.plot(x,y3,color = 'red')
plt.legend([y1_plot,y2_plot],['Integrand','Proposal Density'])
                            

```
